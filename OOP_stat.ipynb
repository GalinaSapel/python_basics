```python
class Statistica:
    def __init__ (self, table: list):
        self.table = table

#min, max, mean, sum, count, медиана и мода
    def mini (self, arr = None):
        if arr is None:
            arr = self.table
        if len(arr) == 1:
            return arr[0]
        else:
            min_of_rest = self.mini(arr[1:])
            if self.table[0] <  min_of_rest:
                return arr[0]
            else:
                return min_of_rest 
        
    def maxi(self, arr = None):
        if arr is None:
            arr = self.table
        if len(arr) == 1:
            return arr[0]
        else:
            max_of_rest = self.maxi(arr[1:])
            if arr[0]> max_of_rest:
                return arr[0]
            else:
                return max_of_rest
    
    def summa(self, arr = None):
        if arr is None:
            arr = self.table
        if len(arr) == 0:
            return 0
        return arr[0] + self.summa(arr[1:])

    def county(self):
        return len(self.table)
    
    def mean_here(self):
        return self.summa()/len(self.table)

    def mediana (self):
        if len(self.table) == 0:
            return None
        sorted_table = sorted(self.table)
        n = len(sorted_table)
        if n%2 == 0:
            return (sorted_table[n//2 - 1] + sorted_table[n//2])/2
        else:
            return sorted_table[n//2]
    
    

    def all_county(self):
        table_dict = {}
        for i in self.table:
            if i in table_dict:
                table_dict[i] += 1
            else:
                table_dict[i] = 1
        return table_dict
    
    def moda(self):
        if len(self.table)==0:
            return None
        all_numbers = self.all_county()
        maximum = self.maxi (list(all_numbers.values()))
        for key, val in all_numbers.items():
            if val == maximum:
                return key
```

```python
table_example1 = Statistica([1, 4, 7, 8, 10, 90])
table_example2 = Statistica([-1, 89, 3])
```

```python
print("Минимум:", table_example1.mini())     
print("Максимум:", table_example1.maxi())    
print("Сумма:", table_example1.summa())      
print("Среднее:", table_example1.mean_here())     
print("Количество:", table_example1.county()) 
print("Подсчет значений:", table_example1.all_county()) 
print("Медиана:", table_example1.mediana())  
print("Мода:", table_example1.moda())        

print("\nДля второго списка:")
print("Минимум:", table_example2.mini())     
print("Максимум:", table_example2.maxi())    
print("Сумма:", table_example2.summa())      
print("Среднее:", table_example2.mean_here())     
print("Количество:", table_example2.county()) 
print("Подсчет значений:", table_example2.all_county()) 
print("Медиана:", table_example2.mediana())  
print("Мода:", table_example2.moda()) 
```
